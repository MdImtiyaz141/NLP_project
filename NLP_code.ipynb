```python
from pathlib import Path
import re
import glob

stop_words_total = []
dir_path = Path("C:/Users/IMTIYAZ/NLP_project/stop_words")

# Iterate over all .txt files in the specified directory
for file in glob.iglob('*.txt', root_dir=dir_path):
    path = dir_path / file
    with open(path, 'r') as f:
        text = f.read()
    
    # Split the text on either '\n' or '|'
    words = re.split(r'\n|\|', text)
    
    # Add the words to the list, excluding empty strings
    stop_words_total.extend(word for word in words if word.strip())

print(stop_words_total)
```

```python
stop_words_total = pd.Series(stop_words_total).str.replace(' ','').str.lower().str.strip().tolist()
```

```python
stop_words_total
```

```python
def remove_stop_words(text):
    new_text = []
    for word in text.split(' '):
        #print('loop',word)
        if word.lower().strip() not in stop_words_total:
            #print('lower',word)
            new_text.append(word)
    return ' '.join(new_text)

# Example usage
cleaned_text = remove_stop_words(df['content'][0])
print(cleaned_text)
```

```python
def postive_score_func(tokens):
    pos_score = 0
    for word in tokens:
        if word in d['positive_words']:
            pos_score += 1
    return pos_score
```

```python
def negative_score_func(tokens):
    neg_score = 0
    for word in tokens:
        if word in d['negative_words']:
            neg_score += 1
    return neg_score
```

```python
df['POSITIVE_SCORE'] = df['tokens'].apply(postive_score_func)
```

```python
df['NEGATIVE_SCORE'] = df['tokens'].apply(negative_score_func)
```

```python
def syllable_count_words(word):
    vowels = 'aeiouAEIOU'
    word = word.lower()
    
    if word.endswith('ed') or word.endswith('es'):
        if len(word)>2 and word[-3] not in vowels:
            return max(1, sum(1 for char in word if char in vowels)-1)
    syllable_count = sum(1 for char in word if char in vowels)
    
    return max(1, syllable_count)
```

```python
df['complex_words'] = df['tokens'].apply(lambda words:np.array(list(syllable_count_words(word) for word in words)))
```

```python
import re
def count_personal_pronoun(text):
    pattern = r'\b(I|we|my|ours|us)\b'
    matches = re.findall(pattern, text, re.IGNORECASE)
    
    matches = [match for match in matches if match.islower() != 'us' or match.isupper()==False]
    
    return len(matches)
```

```python
data['PERSONAL_PRONOUN_COUNT'] = data['content'].apply(count_personal_pronoun)
```

```python
def calculate_average_word_length(text):
    words = text.split()  # Split text into words
    total_words = len(words)  # Total number of words
    total_characters = sum(len(word) for word in words)  # Sum of characters in each word

    # Avoid division by zero
    if total_words == 0:
        return 0
    
    # Calculate average word length
    average_word_length = total_characters / total_words
    
    return average_word_length
```

```python
new_df['AVERAGE_WORD_LENGTH'] = new_df['content_x'].apply(calculate_average_word_length)
```
